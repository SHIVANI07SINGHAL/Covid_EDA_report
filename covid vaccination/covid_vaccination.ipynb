```python
#Renaming the Columns

df = df.rename(columns={df.columns[0]:'Date',
          df.columns[7]:'Male',
          df.columns[8]:'Female',
          df.columns[9]:'Transgender',
          df.columns[10]:'Covaxin',
          df.columns[11]:'CoviShield'})
df.head()
```

```python
#Plotting the Bar Graph 
df1 = df.head()
plt.bar(df1.Date,df1.Covaxin,color=['b'])
plt.xlabel('Date')
plt.ylabel('Covaxin Doses')
plt.title('Bar Graph')
plt.show()
```
